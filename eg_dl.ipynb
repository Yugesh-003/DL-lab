#### Example Code 1: Binary Classification with 1 hidden layer

```python
#Importing libraries
from keras.models import Sequential
from keras.layers import Dense
```

```python
#Define the model
model = Sequential()
model.add(Dense(10, activation='relu', input_shape=(5,))) # Hidden layer with 10 neurons
model.add(Dense(1, activation='sigmoid')) # Output layer
```

```python
#Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
#Print the model summary
model.summary()
```

#### **Example Code 3: Multi class Classification with 1 hidden layer**

```python
# Importing libraries
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
```

```python
# Example dataset
X = np.random.rand(100, 10) # 100 samples, 10 features
y = np.random.randint(0, 4, 100) # 100 labels, 4 classes (0,1,2,3)
```

```python
y
```

```python
# One-hot encoding
y = to_categorical(y, num_classes=4)
```

```python
y
```

```python
# Define model
model = Sequential()
model.add(Dense(16, activation='relu', input_shape=(10,))) # Hidden layer 1
model.add(Dense(8, activation='relu')) # Hidden layer 2
model.add(Dense(4, activation='softmax')) # 4 neurons in output layer to classify 4 classes

### there are 2 hidden layers.. but the question says 1 hidden layer
```

```python
# Define model
model = Sequential()
model.add(Dense(8, activation='relu', input_shape=(10,))) # Hidden layer 1
model.add(Dense(4, activation='softmax')) # 4 neurons in output layer to classify 4 classes
```

```python
# Compile the model
model.compile( optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
model.fit(X, y, epochs=20, batch_size=8)

# for 100 samples of data
# 100 / 8 = 12.5 → rounded up → 13 batches
```

```python
# Evaluate the model
loss, acc = model.evaluate(X, y)
print("Accuracy:", acc)
```

#### **Example Code 4: Regression with 1 hidden layer**

```python
# Importing libraries
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Dummy dataset (100 samples, 3 features)
X = np.random.rand(100, 3) # Creates a dataset with 100 samples (rows)
# Each sample has 3 input features (columns)
# Values are random numbers between 0 and 1
```

```python
y = X[:, 0] * 5 + X[:, 1] * 3 + X[:, 2] * 2 # a simple linear target


# if X[0] = [0.20, 0.50, 0.10], what is the ‘y’ value.
Xs = [0.20, 0.50, 0.10]
ys = 0.20 * 5 + 0.50 * 3 + 0.10 * 2
print("y = ", ys)
```

```python
y = X[:, 0] * 5 + X[:, 1] * 11 + X[:, 2] * 7 # a simple linear target

# whatever feature that is multiplied with highest number has more linarity than the other features
```

```python
X.shape[1]
```

```python
# Plotting each feature against the target
for col in range(0,X.shape[1]):
    plt.figure()
    plt.scatter(X[:,col], y)
```

```python
# Create a simple regression model
model = Sequential()
```

```python
# Input layer + Hidden layer
model.add(Dense(8, activation='relu', input_shape=(3,)))

# Output layer (1 neuron for regression)
model.add(Dense(1, activation='linear'))
```

```python
# Compile the model
model.compile(
optimizer='adam',
loss='mean_squared_error', # MSE is commonly used for regression
metrics=['mae'] # MAE is also a regression metric
)
```

```python
# Train the model
model.fit(X, y, epochs=50, batch_size=8)
```

```python
# Evaluate the model
loss, mae = model.evaluate(X, y)
print("MAE:", mae)
```

```python
# Predict
prediction = model.predict(np.array([[0.2, 0.5, 0.1]]))
print("Prediction:", prediction)
```

#### **Example Code 5 : Regression with 1 hidden layer and tanh as an activation function**

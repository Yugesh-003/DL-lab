```python
#importing libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.regularizers import l1,l2
```

```python
#import dataset
df=pd.read_csv('dataset/indian_houseprice_dataset.csv')

df.info()
```

```python
x=df[['Living Area','Bathrooms','Bedrooms','Lot Size','Age','Fireplace']].values
y=df[['Price']].values
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
ss=StandardScaler()
X_train=ss.fit_transform(X_train)
X_test=ss.transform(X_test)
```

```python
model=Sequential([
    Dense(64,activation='relu',kernel_regularizer=l2(0.01),input_shape=(X_train.shape[1],)),
    Dense(32,activation='relu',kernel_regularizer=l2(0.01)),
    Dense(1,kernel_regularizer=l2(0.01))
])
```

```python
model.compile(optimizer='adam',loss='mse',metrics=['mse'])
```

```python
history=model.fit(X_train,y_train,epochs=300,batch_size=32,
                  validation_split=0.2,verbose=1)
```

```python
#to find loss & mean absolute error values
loss,mae=model.evaluate(X_test,y_test,verbose=0)
```

```python
print(f"Test Loss:{loss:.4f},Test MAE:{mae:.4f}")

```

```python
predication=model.predict(X_test)
print("Sample Predications:",predication[:5].flatten())

```

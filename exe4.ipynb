```python
#importing lbraries
import pandas as pd
import numpy as np
from tensorflow import keras
import tensorflow as tf
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
#importing data
df = pd.read_csv('Student_Performance (1).csv')
```

```python
df.head()
```

```python
df.info()
```

```python
le = LabelEncoder()
df['Extracurricular Activities'] = le.fit_transform(df['Extracurricular Activities'])
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
# Assigning dependent and independent variables
X = df.drop('Performance Index', axis=1)
y = df['Performance Index']
```

```python
# to Standarize data
scalar = StandardScaler()
X_scaled = scalar.fit_transform(X)
```

```python
X_train,X_test, y_train,y_test = train_test_split(X_scaled,y,test_size=0.1,random_state=42)
```

```python
#defining the model
model = keras.Sequential()

model.add(keras.layers.Dense(32, input_shape=(X_train.shape[1],), activation='relu'))
model.add(keras.layers.Dense(16, activation='relu'))
model.add(keras.layers.Dense(8, activation='relu'))
model.add(keras.layers.Dense(1, activation='linear'))
```

```python
model.compile(optimizer=keras.optimizers.SGD(learning_rate=0.001),loss='mean_squared_error')
```

```python
model.fit(X_train, y_train, epochs=100, batch_size=32)
```

```python
test_loss = model.evaluate(X_test, y_test)

# Print the test loss
print(f'Test Loss: {test_loss}')
```

```python
# Use the model to predict Sales for the test data
y_pred = model.predict(X_test)
# Calculate R² score (closer to 1 means better fit)
r2 = r2_score(y_test, y_pred)
print(f'R-squared Score: {r2}')
```

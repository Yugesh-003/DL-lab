```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize

# Download required resources (only once)
nltk.download('punkt_tab')
```

```python

# Input text
text = "I love Data Science. It's intresting"

# Split into sentences
# Sentence tokenize
# sent_tokenize() -> It automatically detects sentence boundaries (like . ! ?)
# and splits a block of text into a list of individual sentences.
sentences = sent_tokenize(text)
print(sentences)
```

```python
print("Sentences:")
# enumerate() adds a counter (index) to an iterable (like a list).

for i, sentence in enumerate(sentences, 1):
    print(f"{i}: {sentence}")
```

```python

# Split into words
words = word_tokenize(text)
print("\nWords:")
print(words)
```

# Simple Deep Learning with Neural Networks

## 1. Importing necessary libraries


```python
import tensorflow as tf
# TensorFlow is used for:  1. Building neural networks 2. Training models  3. Working with tensors (multi-dimensional data)
import numpy as np
from tensorflow import keras
# Keras is a high-level API that makes building neural networks simple.
```

## 2. Define and Compile the Neural Network

```python
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])

# Sequential means the model is built layer by layer, in a sequence.
```

```python
####### 2.2 Compiling the neural network #######

model.compile(optimizer='sgd', loss='mean_squared_error')
```

## 3. Providing the Data

```python
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-2.0, 1.0, 4.0, 7.0, 10.0, 13.0], dtype=float)
```

# 4. Training the Neural Network

```python
model.fit(xs, ys, epochs=500)
# One epoch means the model looks at all your training data one time.
```

# 5. Predicting the output of the Neural Network


```python
print(model.predict([10.0]))
```

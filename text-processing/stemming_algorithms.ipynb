# Stemming Algorithms Examples
This notebook demonstrates different stemming techniques in Python.

```python
# Install NLTK if not already installed
!pip install nltk
```

```python
from nltk.stem import PorterStemmer, LancasterStemmer, SnowballStemmer
import re

words = ["running", "flies", "happiness", "studies", "quickly", "played", "cats"]
```

## Porter Stemmer

```python
ps = PorterStemmer()
for word in words:
    print(word, "->", ps.stem(word))
```

##  Lancaster Stemmer or the Paice-Husk Stemmer

```python
ls = LancasterStemmer()
for word in words:
    print(word, "->", ls.stem(word))
```

## Snowball Stemmer

```python
ss = SnowballStemmer("english")
for word in words:
    print(word, "->", ss.stem(word))
```

## Regex-Based Stemmer

```python
def regex_stem(word):
    return re.sub(r'(ing|ed|ly|s)$', '', word)

for word in words:
    print(word, "->", regex_stem(word))
```

```python
from nltk.stem import RegexpStemmer
reg_stemmer=RegexpStemmer('ing$|s$|e$|able$', min=4)
for word in words:
    print(word, "->", reg_stemmer.stem(word))
```

## Truncation Stemmer

```python
def truncation_stem(word, length=4):
    return word[:length]

for word in words:
    print(word, "->", truncation_stem(word))
```

## Hybrid Stemmer

```python
def hybrid_stem(word):
    ps = PorterStemmer()
    stemmed = ps.stem(word)
    stemmed = re.sub(r'(ly|ness)$', '', stemmed)
    return stemmed

for word in words:
    print(word, "->", hybrid_stem(word))
```

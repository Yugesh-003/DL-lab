## Stopwords

Stop words are common words in a language (like is, the, and, in, to) that are usually removed during text processing because they don’t add significant meaning to the sentence.

They are removed in NLP to reduce data size and focus on the important words.

```python
# Import necessary libraries
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
# Download required resources (only need to run once)
nltk.download('punkt_tab')  #  a pretrained unsupervised model used for tokenizing text into sentences.
nltk.download('stopwords')
```

```python
# Input text
text = "Natural Language Processing helps to get advance knowledge in developing AI models."

# Tokenize text into words
words = word_tokenize(text)

# Load English stopwords
stop_words = set(stopwords.words('english'))
# write a code without using set method and print the stop words
# Remove stopwords
filtered_words = [word for word in words if word.lower() not in stop_words]

# Output the filtered words
print("Original text:")
print(text)
print("\nText after removing stopwords:")
print(" ".join(filtered_words))
```

```python
# Example sentences
examples = [
    "The students from CS(DSA) are so vibrant.",
    "The weather today is very sunny and pleasant.",
    "We are learning Natural Language Processing using Python.",
    "Naresh is not only intelligent but also hardworking.",
    "Can you help me with this task?"
]

# Process each sentence
for i, text in enumerate(examples, 1):
    # Tokenize and remove stopwords
    words = word_tokenize(text)
    filtered = [word for word in words if word.lower() not in stop_words]

    # Display result
    print(f"\nExample {i}:")
    print("Original:", text)
    print("Without Stopwords:", " ".join(filtered))
```

```python
from nltk.tokenize import word_tokenize

# Tamil stopwords list (custom)
tamil_stopwords = [
    "என்று", "இது", "அவன்", "அவள்", "அவர்", "அவர்கள்", "இவர்", "நான்",
    "நீ", "நாம்", "நீங்கள்", "எங்கள்", "உங்கள்", "ஒரு", "இந்த", "அந்த",
    "என்", "உன்", "மற்றும்", "மட்டும்", "உடன்", "மேலும்", "ஆக", "இங்கே",
    "அங்கே", "எப்போது", "எவ்வாறு", "எங்கு", "என்ன", "ஏன்"
]

# Sample Tamil sentences
sentences = [
    "நான் ஒரு புத்தகம் வாசித்தேன்.",
    "அவர் இன்று பள்ளிக்குச் சென்றார்.",
    "இந்த படம் மிகவும் அழகாக உள்ளது.",
    "அவள் என்னுடன் விளையாட விரும்புகிறாள்.",
    "நீங்கள் எப்போது வீட்டிற்கு வருவீர்கள்?"
]

# Stopword removal
for i, text in enumerate(sentences, 1):
    words = word_tokenize(text)
    filtered = [word for word in words if word not in tamil_stopwords]

    print(f"\nExample {i}:")
    print("Original:          ", text)
    print("Without Stopwords: ", " ".join(filtered))
```

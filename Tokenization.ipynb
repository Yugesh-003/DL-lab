```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize

# Download required resources (only once)
nltk.download('punkt_tab')

# Input text
text = "Hello! We are from III B.Sc CS(DSA). We are learning NLP. It's really interesting."

# Split into sentences
# sent_tokenize() -> It automatically detects sentence boundaries (like . ! ?)
# and splits a block of text into a list of individual sentences.
sentences = sent_tokenize(text)
print(sentences)

print("Sentences:")
# enumerate() adds a counter (index) to an iterable (like a list).

for i, sentence in enumerate(sentences, 1):
    print(f"{i}: {sentence}")

# Split into words
words = word_tokenize(text)
print("\nWords:")
print(words)
```

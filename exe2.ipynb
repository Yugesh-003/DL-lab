# Simple Neural Network - Logical AND operation

## Import necessary Libraries

```python
from tensorflow import keras
from tensorflow.keras import layers
# It  is used to import the layers module from TensorFlow's Keras API.

import numpy as np
```

## Define the dataset

```python
# Simple dataset (X: inputs, y: targets)
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])  # Input data (AND problem)
y = np.array([[0], [0], [0], [1]])  # Target outputs
```

## Define the model

```python

model = keras.Sequential([
    layers.Dense(2, activation='relu', input_shape=(2,)),
    layers.Dense(1, activation='sigmoid')
])
```

```python
# Compile the model with optimizer and loss function
model.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['accuracy'])
```

## Train the model

```python
# Train the model
model.fit(X, y, epochs=1000)  # Train for 1000 epochs
```

## Evaluate the model

```python
# Evaluate the model
print("Predictions:")
print(model.predict(X).round())  # Show rounded predictions
```
